```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from copy import deepcopy as copy
%matplotlib inline
```

```python
# Graph1: cnr-2000
vectorizedGraph1 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfFirstMove_cnr-2000.csv')
legacyGraph1 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfLegacyFirstMove_cnr-2000.csv')
# Graph2: eu-2005
vectorizedGraph2 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfFirstMove_eu-2005.csv')
legacyGraph2 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfLegacyFirstMove_eu-2005.csv')
# Graph3: cage15
vectorizedGraph3 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfFirstMove_cage15.csv')
legacyGraph3 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfLegacyFirstMove_cage15.csv')
# Graph4: 333SP
vectorizedGraph4 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfFirstMove_333SP.csv')
legacyGraph4 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfLegacyFirstMove_333SP.csv')
# Graph5: AS365
vectorizedGraph5 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfFirstMove_AS365.csv')
legacyGraph5 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfLegacyFirstMove_AS365.csv')
# Graph6: M6
vectorizedGraph6 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfFirstMove_M6.csv')
legacyGraph6 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfLegacyFirstMove_M6.csv')
# Graph7: NACA0015
vectorizedGraph7 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfFirstMove_NACA0015.csv')
legacyGraph7 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfLegacyFirstMove_NACA0015.csv')
# Graph8: NLR
vectorizedGraph8 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfFirstMove_NLR.csv')
legacyGraph8 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfLegacyFirstMove_NLR.csv')
# Graph9: road_central
vectorizedGraph9 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfFirstMove_road_central.csv')
legacyGraph9 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfLegacyFirstMove_road_central.csv')
# Graph10: road_usa
vectorizedGraph10 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfFirstMove_road_usa.csv')
legacyGraph10 = pd.read_csv('TestPLMFromNetworkIt/AverageTimeOfFirstMove/AverageTimeOfLegacyFirstMove_road_usa.csv')
```

```python
vectorizedGraph = [vectorizedGraph1, vectorizedGraph2, vectorizedGraph3, vectorizedGraph4, vectorizedGraph5, vectorizedGraph6, vectorizedGraph7, vectorizedGraph8, vectorizedGraph9, vectorizedGraph10]
legacyGraph = [legacyGraph1, legacyGraph2, legacyGraph3, legacyGraph4, legacyGraph5, legacyGraph6, legacyGraph7, legacyGraph8, legacyGraph9, legacyGraph10]
graphs = []
for i in range (len(legacyGraph)):
    _vectorizedGraph = vectorizedGraph[i]
    _legacyGraph = legacyGraph[i]
    _vectorizedGraph = _vectorizedGraph.loc[_vectorizedGraph['Time'] >0]
    _legacyGraph = _legacyGraph.loc[_legacyGraph['Time'] >0]
    graph = copy(_vectorizedGraph[['Time', 'Number of Iterations']])
    graph.columns = ['Vectorized Time', 'Vectorized Number of Iterations']
    graph['Legacy Time'] = _legacyGraph['Time']
    graph['Legacy Number of Iterations'] = _legacyGraph['Number of Iterations']
    graphs.append(graph)
```

## Graph: cnr-2000

```python
graphs[0][['Vectorized Time', 'Legacy Time']].plot(kind='bar', title="Average Time per Move in Sorted VS Legacy Graph-cnr-2000", figsize=(15,7))
```

```python
graphs[0][['Vectorized Number of Iterations', 'Legacy Number of Iterations']].plot(kind='bar', title="Number of Iteration per Move in Sorted VS Legacy Graph-cnr-2000", figsize=(15,7))
```
